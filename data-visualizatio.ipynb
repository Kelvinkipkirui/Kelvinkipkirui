```python
import matplotlib.pyplot as pl
import seaborn as sns
%matplotlib inline
```

```python
crop=[34,45,67,89,67,89]
```

```python
pl.plot(crop);
yea=[1960,1970,1980,1990,2000,2010]
```

```python

pl.plot(yea,crop)
pl.ylabel('yield_in_tons')
pl.xlabel("mwaka_1");
```

```python
cows=[21,56,78,98,99]
goats=[23,30,35,40,56]
year=[2011,2012,2013,2014,2015]
```

```python
pl.plot(year,cows,marker="o",c="red",lw=3,ls="-",ms=7)
pl.plot(year,goats,marker="o",c="blue",lw=3,ls="-",ms=7)
pl.xlabel("average weight")
pl.ylabel("year")
pl.title("kericho yield")
pl.legend(["cows","goats"]);
```

```python
pl.plot(year,cows,marker="o",c="red",lw=3,ls="--",ms=7,alpha=.5)# ls means linestyle, lw means line width, ms means marker size alpha means optic of line
```

```python
pl.plot(year,cows,"o--b",lw=3,alpha=1)#o means marker, -- means linestyle, b means color
pl.legend("cow")
pl.title("Kenya")
pl.xlabel("year")
pl.ylabel("weight");
```

```python
pl.plot(year,cows,"ob",lw=3,alpha=1)#no line
```

```python
pl.plot(year,cows,"or",lw=3,alpha=1)
pl.figure(figsize=[7,10]);#plot size
```

```python
pl.plot(year,cows,"o--b",lw=3,alpha=1)#o means marker, -- means linestyle, b means color
pl.legend("cow")
pl.title("Kenya")
pl.xlabel("year")
pl.ylabel("weight")
sns.set_style("whitegrid");
```
